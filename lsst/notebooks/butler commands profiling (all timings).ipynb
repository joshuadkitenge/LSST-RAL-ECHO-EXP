```python
# Imports 
import pandas as pd
import numpy as np
import pylab as plt
from scipy import stats
from matplotlib import pyplot as plt
```

```python
def profilingtimings(cpu_mem_path,command):
    
    # loading the data from the txt files
       
    cpu_mem_col2 = np.loadtxt(cpu_mem_path,skiprows=0,usecols=1,dtype=str)
    cpu_mem_col3 = np.loadtxt(cpu_mem_path,skiprows=0,usecols=3,dtype=str)
    
    # Splicing the data to get the vprrect data pint from the output text and convertvalues into floats
    
    ip=np.where(cpu_mem_col2=="python")[0]
        
    CPU = list(map(float, cpu_mem_col3[ip]))

    return len(CPU)
    
    
    
    
```

```python
# function that create the paths to the txt file 

def path(storage,command):
    # edit "/home/vrs42921" to your homespace 
    cpu = "/home/vrs42921/LSST-RAL-ECHO-EXP/lsst/time_test/{}/{}/info_cpu_mem.txt" .format(storage,command)
    return cpu

# command line task / pipetasks

com=["create","register-instrument","import","ingest-raws","define-visit","write-curated-calibrations",
     "processccd","make-discrete-skymap","coaddtions","multiband"]

# end point storages

store =["Local","Local_cd","S3","S3_cd","cephfs","cephfs_cd","cephfs_reglocal","cephfs_reglocal_cd"]

# print formatting 

class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
   NEXT ='\n'
```

```python
# collating all the data into a list 
col1= []
for i in range(len(store)):
    for j in range(len(com)):
        times = profilingtimings(path(store[i],com[j]),com[j])
        col1.append(times)

        
print(len(col1))
```
