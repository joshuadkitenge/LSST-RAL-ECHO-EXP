## Commands

Available Commands:<br><br>
  - about           :Get quota information from the remote.<br><br>
  - authorize       :Remote authorization.<br><br>
  - backend         :Run a backend specific command.<br><br>
  - cat             :Concatenates any files and sends them to stdout.<br><br>
  - check           :Checks the files in the source and destination match.<br><br>
  - cleanup         :Clean up the remote if possible.<br><br>
  - config          :Enter an interactive configuration session.<br><br>
  - copy            :Copy files from source to dest, skipping already copied.<br><br>
  - copyto          :Copy files from source to dest, skipping already copied.<br><br>
  - copyurl         :Copy url content to dest.<br><br>
  - cryptcheck      :Cryptcheck checks the integrity of a crypted remote.<br><br>
  - cryptdecode     :Cryptdecode returns unencrypted file names.<br><br>
  - dedupe          :Interactively find duplicate filenames and delete/rename them.<br><br>
  - delete          :Remove the files in path.<br><br>
  - deletefile      :Remove a single file from remote.<br><br>
  - genautocomplete :Output completion script for a given shell.<br><br>
  - gendocs         :Output markdown docs for rclone to the directory supplied.<br><br>
  - hashsum         :Produces a hashsum file for all the objects in the path.<br><br>
  - help            :Show help for rclone commands, flags and backends.<br><br>
  - link            :Generate public link to file/folder.<br><br>
  - listremotes     :List all the remotes in the config file.<br><br>
  - ls              :List the objects in the path with size and path.<br><br>
  - lsd             :List all directories/containers/buckets in the path.<br><br>
  - lsf             :List directories and objects in remote:path formatted for parsing.<br><br>
  - lsjson          :List directories and objects in the path in JSON format.<br><br>
  - lsl             :List the objects in path with modification time, size and path.<br><br>
  - md5sum          :Produces an md5sum file for all the objects in the path.<br><br>
  - mkdir           :Make the path if it doesn't already exist.<br><br>
  - mount           :Mount the remote as file system on a mountpoint.<br><br>
  - move            :Move files from source to dest.<br><br>
  - moveto          :Move file or directory from source to dest.<br><br>
  - ncdu            :Explore a remote with a text based user interface.<br><br>
  - obscure         :Obscure password for use in the rclone config file.<br><br>
  - purge           :Remove the path and all of its contents. <br><br>
  - rc              :Run a command against a running rclone. <br><br>
  - rcat            :Copies standard input to file on remote.<br><br>
  - rcd             :Run rclone listening to remote control commands only.<br><br>
  - rmdir           :Remove the empty directory at path.<br><br>
  - rmdirs          :Remove empty directories under the path.<br><br>
  - selfupdate      :Update the rclone binary.<br><br>
  - serve           :Serve a remote over a protocol.<br><br>
  - settier         :Changes storage class/tier of objects in remote.<br><br>
  - sha1sum         :Produces an sha1sum file for all the objects in the path.<br><br>
  - size            :Prints the total size and number of objects in remote:path.<br><br>
  - sync            :Make source and dest identical, modifying destination only.<br><br>
  - test            :Run a test command<br><br>
  - touch           :Create new file or change file modification time.<br><br>
  - tree            :List the contents of the remote in a tree like fashion.<br><br>
  - version         :Show the version number.<br><br>

## How to
        
        
### Make a bucket

» rclone mkdir {name_config}:{insert_name_of_the_new_bucket}  
- e.g. » rclone mkdir remote:mybucket/mybucket_folder <br><br>
- The {name_config} is the name you choose in the configuration <br><br>

###  List all buckets

» rclone lsd {name_config}: 
- e.g. » rclone lsd remote:<br>
     -1 2021-05-06 12:00:26        -1 mybucket
     
### Copy files into a bucket 
» rclone copy {file_or_directory_name} {name_config}:{name_of_bucket}
- e.g. » rclone copy kidneybeans.py remote:mybucket {file} <br>
- e.g. » rclone copy beans remote:mybucket {directory} <br>
- Ps. you have to been in the same directory as file or directory <br>
- {directory only} This will copy all of the items in the beans directory into mybucket bucket 

### List all files within a bucket

» rclone ls {name_config}:{name_of_bucket}<br>
    
- e.g. » rclone ls remote:mybucket <br>
    {size} {file_name}<br>
    100 kidneybeans.py <br>
    123 pintobeans.py <br>
    200 cranberrybeans.py <br>
    234 flatbeans.py <br>

### Copy files from a bucket into your directory
» rclone copyto {name_config}:{name_of_bucket} {destination}
- e.g. » rclone copyto remote:mybucket /home/gtf74638/mybucket <br>
- This will copy all the content to yourlocal directory in a folder called mybucket 

### Create a url link for your bucket
» rclone link {name_config}:{name_of_bucket}/{path_to_file}
- e.g. » rclone link - expire 1d remote:mybucket/kidneybean.py <br>
-  The expire command limits the amount of time that the link is active

### Delete a file within a bucket
» rclone deletefiles {name_config}:{name_of_bucket}/{path_to_file}
- e.g » rclone deletefiles remote:mybucket/kidneybeans.py

### Delete all files from bucket
» rclone delete {name_config}:{name_of_bucket}
- e.g. » rclone delete remote:mybucket

### Delete a bucket
» rclone rmdir {name_config}:{name_of_bucket}
- e.g. rclone rmdir remote:mybucket
- Make sure you empty the bucket before deleting
- To delete mulitple empty bucket use "rmdirs" instead of "rmdir"
